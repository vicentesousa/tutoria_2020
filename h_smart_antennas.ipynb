## Hands-on Smart Antennas

**Objetivos:**
- Apresentar o que são antenas inteligentes;
- Apresentar como calcular o ganho de algumas estratégias de antenas inteligentes;
- Apresentar como antenas inteligentes pode ser usadas para aumentar a capacidade de um sistema.

## Introdução

A grande comodidade e rapidez que a comunicação móvel proporciona é um fator chave para explicar o fenomenal aumento na procura por esses serviços. Esse fato exige do setor tecnológico um esforço técnico e científico a fim de melhorar os sistemas de forma a atender essa demanda.

Alocação de mais freqüência, novas técnicas de múltiplo acesso, controle de potência, melhores codificadores de voz (vocoders), adição de microcélulas (cell splintting), setorização e antenas inteligentes são exemplos de opções para o aumento da capacidade das redes de telefonia móvel. Contudo, algumas dessas opções apresentam algumas limitações que, às vezes, as tornam inviáveis.

A introdução de novas frequências, por exemplo, mmWave (30 GHz a 300 GHz), são vantajosas por alocarem maior banda. Porém, aparecem várias desvantagens em relação aos sistemas com frequências no espectro de microondas. A perda de percurso é maior em portadoras mmWaves, gerando um problema de cobertura, necessitando de mais estações.
   
A adição de microcélulas acarreta um aumento efetivo no número de usuários por área. Contudo, a sua implantação requer um novo planejamento da rede, alocação de espaço para instalação de novas antenas e mudança no gerenciamento da rede, incluindo nova política de handoff. 

Com a setorização, a antena omnidirecional da estação radio base (ERB) é substituída por antenas direcionais, com intuito de reduzir a interferência co-canal e consequentemente aumentar a SIR (signal-interference ratio). Essa ação pode diminuir o perfil de interferência de forma a permitir a redução de padrão de reuso. Com a redução do padrão de reuso, o número de canais por área aumenta. Entretanto, o número total de canais disponível na ERB é dividido pelo número de setores, gerando uma perda de eficiência de tráfego na célula, que, algumas vezes, não pode ser superada pela redução do padrão de reuso. Além desse problema, o tamanho de um setor é bem menor que o tamanho de uma célula, acarretando um aumento no número de handoffs.


As antenas inteligentes aparecem como um grande aliado entre as técnicas de aumento de capacidade do sistema de comunicação móvel, pois possibilita ganhos expressivos de diversidade e de multiplexação espacial. Diversidade é uma técnica que combina as múltiplas cópias do sinal (devido aos multipercursos ou a múltiplas antenas), de forma a melhorar a SINR (signal-interference plus noise ratio) do sinal recebido. Essa técnica é muito utilizada atualmente e é aproveitada tanto para combater o desvanecimento de larga escala quanto o desvanecimento de pequena escala. Já a multiplexação espacial usa as múltiplas antenas para transmistir informações em paralelo, aumentando a taxa de transmissão.

## Técnicas Clássicas de Diversidade


Diversidade é um método usado para recuperar um sinal transmitido sobre independentes caminhos de desvanecimento. O objetivo é combinar múltiplos sinais e reduzir o efeito do desvanecimento.


Quanto ao tipo de desvanecimento que combatem, os esquemas de diversidade se dividem em dois tipos: **esquema de diversidade macroscópica** e **esquema de diversidade microscópica**.

### Diversidade macroscópica

Esse tipo de diversidade visa o combate do desvanecimento de larga escala, isto é, perda de percurso e sombreamento. A justificativa do uso de tal esquema de diversidade reside no fato de que dois desvanecimentos profundos ocorrem raramente no mesmo instante de tempo, como ilustrado na Figura a seguir. Assim, são usadas duas ou mais antenas transmissoras de sites diferentes para transmitir o mesmo sinal, cabendo ao móvel selecionar o melhor entre dois sinais recebidos.

### Diversidade microscópica

Este tipo de diversidade é aplicada em sinais que sofrem o mesmo desvanecimento de larga escala, visando combater o desvanecimento de pequena escala. No esquema de diversidade microscópica, réplicas do sinal são recebidas por um conjunto de antenas em um mesmo site. Essas réplicas são combinadas no receptor gerando o sinal resultante. Dependendo do método de obtenção dessas réplicas, se classificam os métodos de diversidade em vários tipos, entre eles: diversidade espacial, diversidade angular, diversidade por polarização e diversidade temporal.

Na diversidade espacial, réplicas descorrelacionadas do sinal são recebidas por várias antenas adequadamente espaçadas e combinadas no receptor. 

A diversidade angular é obtida por meio de sinais captados por antenas que cobrem setores diferentes. Neste caso, o ganho de diversidade está ligado ao número de setores (ou de feixes) que servem de ramos de diversidade. 

A diversidade por polarização é obtida pela transmissão de cópias do sinal nas componentes vertical e horizontal da antena. Na diversidade de polarização, o espaço entre as antenas não é mais um requisito tão importante quanto na diversidade espacial, mas o ganho de diversidade é reduzido e o número de caminhos de diversidade se reduz a dois. 




## Antenas Inteligentes

Antenas inteligentes são arranjos de antenas separadas espacialmente e conectadas a uma rede de combinação de sinais. Os métodos de combinação podem ser de vários tipos entre eles: combinação por diversidade, formação ótima de feixe e combinação ótima por algoritmos adaptativos.

As antenas inteligentes melhoram a capacidade do sistema devido a customização do padrão de radiação das antenas no transmissor e receptor, os quais podem concentrar ou anular potência em um determinado ponto do espaço. 

Estruturalmente, uma antena Inteligente é composta de um arranjo de sensores e uma unidade de combinação de sinais. Cada sensor capta uma amostra do sinal separado espacialmente, a qual será tratada apropriadamente pelo combinador de sinais, objetivando um ganho de desempenho na comunicação da informação. A Figura a seguir ilustra um arranjo de antenas, em que $h_i(t)$ é a componente de canal entre o i-ésimo sensor e móvel.

```matlab

\section{Estrat\'{e}gias de Antenas do $WISE_{SIM}$}
\label{Sec:SAProp}

Nesta se\c{c}\~{a}o, o leitor ter\'{a} contato com a formula\c{c}\~{a}o dos algoritmos
de antenas inteligentes utilizados nesta disserta\c{c}\~{a}o, bem como
suas vantagens e desvantagens, seja no aspecto construtivo como no
aspecto de desempenho.

\subsection{Antenas Inteligentes}
\subsubsection{Conceitos e Formula\c{c}\~{a}o Geral}

Antena Inteligente \'{e} uma estrutura composta de um arranjo de
sensores e uma unidade de combina\c{c}\~{a}o de sinais. Cada sensor capta
uma amostra do sinal separado espacialmente, a qual ser\'{a} tratada
apropriadamente pelo combinador de sinais, objetivando um ganho de
desempenho na comunica\c{c}\~{a}o da informa\c{c}\~{a}o. A figura \ref{Fig:Array}
ilustra um arranjo de antenas onde $h_{i}(t)$ \'{e} a componente de
canal entre o i-\'{e}simo sensor e m\'{o}vel.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.4]{Array.eps}
\end{center}
\vspace{-3ex} \caption{\label{Fig:Array}\hspace{-0.4em}{\bf:}
Exemplo gen\'{e}rico de antena inteligente usada para recep\c{c}\~{a}o.}
\end{figure}

Utilizando uma nota\c{c}\~{a}o vetorial, pode-se escrever o vetor de
resposta do canal como na equa\c{c}\~{a}o \ref{Eq:VetorCanal}.

\begin{equation}\label{Eq:VetorCanal}
{\bf h}(t) = [\begin{array}{*{20}c}
   {h_0 (t)} & {h_1 (t)} &  \cdots  & {h_{M-1}(t)}]  \\
\end{array}
\end{equation}

\noindent onde $M$ \'{e} o n\'{u}mero de antenas do arranjo. Neste
trabalho, cada sub-canal \'{e} modelado de acordo com a se\c{c}\~{a}o
\ref{Sec:ChannelModel}. Tamb\'{e}m, sem perda de generalidade, para
arranjos de uns poucos comprimentos de onda \'{e} razo\'{a}vel assumir que
a perda de percurso e o sombreamento afetam de forma igual todos
os componentes $h_{i}(t)$ \cite{PhD:Rodrigo99}.

Antes da discuss\~{a}o das solu\c{c}\~{o}es de antenas inteligentes, alguns
par\^{a}metros relacionados \`{a} teoria geral de arranjos de antenas
ser\~{a}o discutidos:

\begin{description}
\item[Diagrama de Radia\c{c}\~{a}o]: diagrama direcional de uma antena,
onde a distribui\c{c}\~{a}o relativa da pot\^{e}ncia radiada \'{e} representada
como fun\c{c}\~{a}o da dire\c{c}\~{a}o espacial. Um exemplo imediato da no\c{c}\~{a}o do
que seja um diagrama de radia\c{c}\~{a}o, \'{e} a zona iluminada da estrada
pelos far\'{o}is de um autom\'{o}vel, como mostra a figura
\ref{Fig:CarroPattern}. O diagrama ou padr\~{a}o de um feixe de
radia\c{c}\~{a}o luminosa mostra, graficamente, onde h\'{a} mais radia\c{c}\~{a}o. No
caso da antena receptora, mostra as dire\c{c}\~{o}es de melhor recep\c{c}\~{a}o de
sinal \cite{Book:Jaroslav87}.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.5]{CarroPattern.eps}
\end{center}
\vspace{-3ex}
\caption{\label{Fig:CarroPattern}\hspace{-0.4em}{\bf:} Exemplo
ilustrativo de um diagrama de  um feixe de radia\c{c}\~{a}o luminosa.}
\end{figure}

O padr\~{a}o de radia\c{c}\~{a}o de um arranjo de antenas pode ser modelado
pelo produto entre o padr\~{a}o de radia\c{c}\~{a}o de um elemento do arranjo
($f_{a}(\theta ,\phi$)) e o fator do arranjo ($F(\theta ,\phi)$),
como mostrado na equa\c{c}\~{a}o \ref{Eq:Radiation}.

\begin{equation}\label{Eq:Radiation}
G(\theta ,\phi) = f_a (\theta ,\phi).F(\theta ,\phi)
\end{equation}

\noindent onde $\theta$ e $\phi$ s\~{a}o os \^{a}ngulos do azimute e de
eleva\c{c}\~{a}o dos elementos do arranjo, respectivamente.

A figura \ref{Fig:Radiation} mostra tr\^{e}s curvas que representam o
padr\~{a}o de radia\c{c}\~{a}o de uma antena do arranjo ($f_{a}(\theta
,\phi$)), o fator do arranjo ($F(\theta ,\phi)$) e o padr\~{a}o de
radia\c{c}\~{a}o do arranjo ($G(\theta ,\phi)$).

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.4]{Radiation.eps}
\end{center}
\vspace{-3ex} \caption{\label{Fig:Radiation}\hspace{-0.4em}{\bf:}
Padr\~{a}o de radia\c{c}\~{a}o para um arranjo linear com uma separa\c{c}\~{a}o entre
antenas de $\lambda$/2 e \^{a}ngulo de eleva\c{c}\~{a}o igual a 0.}
\end{figure}

\item[Fator do Arranjo]: padr\~{a}o de radia\c{c}\~{a}o do campo de um arranjo
de elementos radiantes isotr\'{o}picos, isto \'{e}, que apresentam as
mesmas propriedades f\'{\i}sicas em todas as dire\c{c}\~{o}es.


\item[L\'{o}bulo Principal]: maior por\c{c}\~{a}o do diagrama direcional de
uma antena limitada por um ou dois cones de nulos, isto \'{e}, a
por\c{c}\~{a}o que contem a dire\c{c}\~{a}o de m\'{a}xima pot\^{e}ncia radiada.


\item[L\'{o}bulos Laterais]: l\'{o}bulos direcionados espacialmente de
forma secund\'{a}ria, cujas pot\^{e}ncias s\~{a}o projetadas para serem
inferiores \`{a} pot\^{e}ncia do l\'{o}bulo principal. Esses l\'{o}bulos
representam tamb\'{e}m a parcela de amplifica\c{c}\~{a}o (ou atenua\c{c}\~{a}o) que
ser\'{a} dada aos usu\'{a}rios interferentes.


\item[Largura do Feixe]: espessura angular do l\'{o}bulo principal.
Geralmente, a largura do feixe \'{e} igual \`{a} espessura angular entre
dois pontos do l\'{o}bulo principal que tem pot\^{e}ncia 3dB abaixo do seu
pico m\'{a}ximo.
\end{description}

O termo ``intelig\^{e}ncia" \;, associado \`{a} t\'{e}cnica de
antena inteligente, provem do m\'{e}todo de combina\c{c}\~{a}o dos
sinais oriundos dos sensores do arranjo. Assim, o m\'{e}todo de
combina\c{c}\~{a}o deve ser escolhido dependendo do tipo de
abordagem espa\c{c}o-temporal do canal utilizado. Dependendo da
rela\c{c}\~{a}o entre os sub-canais $h_{i}$, o canal
espa\c{c}o-temporal pode ser tratado por uma abordagem de
diversidade ou de radar. Na abordagem por diversidade, assume-se
que o sistema experimenta forte desvanecimento seletivo no
espa\c{c}o devido ao grande espalhamento angular, implicando em
uma pequena dist\^{a}ncia de coer\^{e}ncia. Isso estabelece que,
deve-se garantir que a dist\^{a}ncia entre os elementos do arranjo
seja maior que a dist\^{a}ncia de coer\^{e}ncia do canal,
resultando em desvanecimentos temporais independentes para cada
sub-canal $h_{i}$.  Dessa forma, t\'{e}cnicas de
combina\c{c}\~{a}o de sinais exploram um ganho por diversidade
espacial, desconsiderando a exist\^{e}ncia de sinais interferentes
(co-canais). Por isso esse m\'{e}todo \'{e} \'{o}timo apenas na
aus\^{e}ncia da interfer\^{e}ncia de m\'{u}ltiplo acesso.
Diferente da abordagem por diversidade, os m\'{e}todos de
combina\c{c}\~{a}o de sinais que se valem da abordagem por radar,
t\^{e}m a preocupa\c{c}\~{a}o de extrair um ganho de cancelamento
dos sinais interferentes indesejados. Essa abordagem \'{e}
caracterizada quando o espalhamento angular \'{e} muito pequeno,
de forma que o sistema experimenta desvanecimento plano no
espa\c{c}o. Devido ao estreito espalhamento angular, o termo
``radar" \; \'{e} caracterizado. Essa associa\c{c}\~{a}o se deve
ao fato de que um pequeno espalhamento angular \'{e}, fisicamente,
associado a fontes emissoras pontuais com linha de visada direta.
Neste trabalho, adotou-se o poderio de cancelamento de
interfer\^{e}ncia disponibilizado na abordagem por radar, onde a
dist\^{a}ncia entre os elementos do arranjo \'{e} menor que a
dist\^{a}ncia de coer\^{e}ncia do canal. Tais suposi\c{c}\~{o}es
implicam que todos os sub-canais $h_{i}$ experimentam o mesmo
desvanecimento r\'{a}pido, de forma que a envolt\'{o}ria dos
sinais em todas as antenas \'{e} essencialmente a mesma, isto
\'{e}, plenamente correlacionada \cite{PhD:Rodrigo99}.

Outro fator de grande import\^{a}ncia no planejamento da aplica\c{c}\~{a}o de
antena inteligente \'{e} o espa\c{c}amento entre as antenas. Ele deve ser
escolhido levando em considera\c{c}\~{a}o tr\^{e}s aspectos:

\begin{itemize}

\item Aspectos construtivos: a dist\^{a}ncia entre as antenas do
arranjo caracterizar\'{a} o tamanho da antena inteligente. A
constru\c{c}\~{a}o do arranjo de antenas pode ser inviabilizado se for
escolhida uma grande dist\^{a}ncia entre as antenas do arranjo,
principalmente se a sua implementa\c{c}\~{a}o for no equipamento do
usu\'{a}rio (celular, \textit{palmtop} ou \textit{laptop}, por
exemplo).

\item Efeito dos l\'{o}bulos laterais (\textit{grating lobes}): em
\cite{Paper:Simon90} \'{e} mostrado que, para arranjos com antenas
separadas de uma dist\^{a}ncia maior que a metade do comprimento de
onda da portadora ($\lambda/2$), a resolu\c{c}\~{a}o espacial alcan\c{c}ada
pelo feixe direcional \'{e} maior, isto \'{e}, se consegue uma menor
largura de feixe. Por\'{e}m, junto com esse benef\'{\i}cio, \'{e} radiada mais
pot\^{e}ncia pelos l\'{o}bulos laterais. Na verdade, o padr\~{a}o de radia\c{c}\~{a}o
ser\'{a} composto de l\'{o}bulos principais secund\'{a}rios como mostra a
figura \ref{Fig:GratingEffect}.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.4]{GratingEffect.eps}
\end{center}
\vspace{-4ex}
\caption{\label{Fig:GratingEffect}\hspace{-0.4em}{\bf:} Efeito dos
l\'{o}bulos laterais: compara\c{c}\~{a}o do padr\~{a}o de radia\c{c}\~{a}o para um arranjo
linear com uma separa\c{c}\~{a}o entre antenas de $\lambda$/2 e
3$\lambda$/2.}
\end{figure}

O requerimento $d<\lambda/2$ tamb\'{e}m pode ser visto como um
analogia espacial ao teorema da amostragem, e o seu descumprimento
acarretar\'{a} em um efeito de \textit{aliasing} espacial
\cite{Book:Haykin02}.

\item Efeito do acoplamento m\'{u}tuo: em \cite{Paper:Inder83} \'{e}
mostrado que o efeito do acoplamento m\'{u}tuo, isto \'{e}, imped\^{a}ncia
m\'{u}tua entre as antenas do arranjo, degrada o desempenho do arranjo
de antenas quando a separa\c{c}\~{a}o espacial entre as antenas \'{e} menor
que $\lambda/2$.
\end{itemize}

Dessa forma, foi escolhido o valor $\lambda/2$ como a dist\^{a}ncia
entre os elementos do arranjo de antenas empregado neste trabalho.
Esse valor estabelece um compromisso entre um menor n\'{\i}vel de
acoplamento m\'{u}tuo e a menor presen\c{c}a de l\'{o}bulos laterais
indesej\'{a}veis.

Considerando a abordagem radar, o canal espa\c{c}o-temporal pode ser
modelado como mostra a equa\c{c}\~{a}o \ref{Eq:CanalST}.

\begin{equation}\label{Eq:CanalST}
    {\bf h}(t) = \alpha (t){\bf f}
\end{equation}

\noindent onde $\alpha$(t) representa o desvanecimento do usu\'{a}rio
em quest\~{a}o e \textbf{f} o vetor de resposta do arranjo.
Considerando que, para um instante de tempo fixo, o desvanecimento
visto por cada antena do arranjo \'{e} o mesmo, deve-se estudar a
teoria de conjunto que rege um conjunto de antenas a fim de
extrair um modelo para o vetor de resposta do arranjo
\cite{Book:Jaroslav87}. Considere uma s\'{e}rie de antenas isotr\'{o}picas
alinhadas uniformemente, vide figura \ref{Fig:ULA}. Essa geometria
de organiza\c{c}\~{a}o de antenas forma um arranjo chamado de ULA
(\textit{Uniform Linear Array}), que ser\'{a} a abordagem geom\'{e}trica
de antenas usada nos algoritmos explorados nesta disserta\c{c}\~{a}o.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.4]{ULA.eps}
\end{center}
\vspace{-4ex} \caption{\label{Fig:ULA}\hspace{-0.4em}{\bf:}
Arranjo uniforme de antenas uniforme no plano do azimute: onda
plana incidente com \^{a}ngulo de chegada $\theta_{n}$.}
\end{figure}
%esta figura esta dando problema na gera\c{c}\~{a}o do PS

N\~{a}o \'{e} dif\'{\i}cil notar que as antenas sucessivas t\^{e}m sua radia\c{c}\~{a}o
adiantada de um \^{a}ngulo de fase que corresponde \`{a} diferen\c{c}a de
dist\^{a}ncia percorrida das ondas que emanam das mesmas. Essa medida
de dist\^{a}ncia \'{e} feita ao longo do percurso de radia\c{c}\~{a}o, com
refer\^{e}ncia a um plano perpendicular (vide figura \ref{Fig:ULA}).
Lembrando que uma defasagem de 2$\pi$ ocorre quando se percorre a
dist\^{a}ncia de um comprimento de onda da portadora, a diferen\c{c}a de
fase entre uma antena $i$ e a primeira antena (antena de
refer\^{e}ncia), pode ser escrita como na equa\c{c}\~{a}o \ref{Eq:DifFase}.

\begin{equation}\label{Eq:DifFase}
    \psi _{i,{\theta_n}}  = \frac{{2\pi{\kern 1pt}i{\kern 1pt}d{\kern 1pt}sen(\theta _n )}}{\lambda
    }, \hspace{1em}i = 0, ..., M-1
\end{equation}

\noindent onde \textit{d} \'{e} a dist\^{a}ncia entre os elementos do
arranjo, $\theta_{n}$ \'{e} o DOA do \textit{n}-\'{e}simo usu\'{a}rio e
$\lambda$ \'{e} o comprimento de onda da portadora. Utilizando essa
abordagem, o vetor que representa a resposta do arranjo de antenas
em cada uma das $M$ antenas \'{e} mostrado na equa\c{c}\~{a}o
\ref{Eq:ArrayReponse}.

\begin{equation}\label{Eq:ArrayReponse}
    {\bf f}(\theta _n ) = {[\begin{array}{*{20}c}
    1 & {e^{j\psi_{1,{\theta_n}}}} & {e^{j\psi _{2,{\theta_n}}}} &  \cdots  & {e^{j\psi _{{M - 1},{\theta_n}}} }  \\
\end{array}]}^T
\end{equation}

A modelagem sist\^{e}mica do canal espacial aqui estudado \'{e} baseada no
conhecimento ideal do \^{a}ngulo de chegada, considerando somente a
posi\c{c}\~{a}o do terminal m\'{o}vel na rede celular. \'{E} razo\'{a}vel adotar essa
modelagem, pois os usu\'{a}rios est\~{a}o distribu\'{\i}dos uniformemente em
toda \'{a}rea de cobertura e procura-se caracterizar um sistema
macrocelular com um pequeno espalhamento angular. Uma descri\c{c}\~{a}o
mais completa de m\'{e}todos de estima\c{c}\~{a}o de DOA pode ser vista em
\cite{Paper:Godara97}. Assim, com a estima\c{c}\~{a}o do \^{a}ngulo de chegada
do usu\'{a}rio desejado pode-se encontrar o valor de $f(\theta_{n})$
de forma direta.

Agora, ser\'{a} apresentada a formula\c{c}\~{a}o das antenas inteligentes em
um sistema com a presen\c{c}a de m\'{u}ltiplos usu\'{a}rios. Considerando $N$
usu\'{a}rios transmitindo concorrentemente no sistema, o vetor de
sinais recebidos no arranjo pode ser escrito como na equa\c{c}\~{a}o
\ref{Eq:SinalRec}.

\begin{equation}\label{Eq:SinalRec}
    {\bf u}(t) = \sum\limits_{n = 0}^{N-1} {{\bf h}_n } s_n (t)
\end{equation}

\noindent onde ${\bf h}_{n}$ \'{e} o canal espa\c{c}o-temporal relativo ao
usu\'{a}rio $n$ e $s_{n}$(t) \'{e} o sinal transmitido pelo n-\'{e}simo
usu\'{a}rio, dado pela equa\c{c}\~{a}o \ref{Eq:PulsoTrans}.

\begin{equation}\label{Eq:PulsoTrans}
    s_n (t) = \sum\limits_{k =  - \infty }^\infty  {\sqrt {P_n } } d_n
    (k)p(t - kT)
\end{equation}

\noindent onde $P_{n}$ \'{e} a pot\^{e}ncia de transmiss\~{a}o, $d_{n}$ \'{e} a
seq\"{u}\^{e}ncia de s\'{\i}mbolos do \textit{n}-\'{e}simo usu\'{a}rio, $p(t)$ \'{e} a
forma do pulso de transmiss\~{a}o e $T$ \'{e} o intervalo de s\'{\i}mbolo.

Ap\'{o}s o processamento que envolve os pesos do arranjo, a sa\'{\i}da da
combina\c{c}\~{a}o de sinais no arranjo de antenas do receptor \'{e} dada pela
equa\c{c}\~{a}o \ref{Eq:SaidaArranjo}.

\begin{equation}\label{Eq:SaidaArranjo}
    z(t) = {\bf w}_n^H {\bf u}(t)
\end{equation}

\noindent onde \textbf{w}=[$w_{0}\;\;w_{1}\;\; \cdots
\;\;w_{M-1}$]$^H$ \'{e} o vetor de pesos que visa tornar eficiente a
recep\c{c}\~{a}o do usu\'{a}rio $n$ e o operador $[\cdot]^H$ significa a
transposta do complexo conjugado (Hermitiano) de um vetor.

Para a constru\c{c}\~{a}o dos algoritmos que ser\~{a}o descritos
posteriormente, utilizou-se a solu\c{c}\~{a}o \'{o}tima de Wiener para
determina\c{c}\~{a}o dos pesos do arranjo. A determina\c{c}\~{a}o dos pesos \'{e}
baseada na considera\c{c}\~{a}o de uma frente de onda plana no azimute.
Dessa forma, a componente espacial do canal de propaga\c{c}\~{a}o depende
apenas da dire\c{c}\~{a}o de chegada (DOA) dos usu\'{a}rios, como mostrado nas
equa\c{c}\~{o}es \ref{Eq:CanalST} e \ref{Eq:ArrayReponse}. Seguindo essas
considera\c{c}\~{o}es, o vetor de pesos do arranjo para o $n$-\'{e}simo
usu\'{a}rio pode ser escrito como na equa\c{c}\~{a}o \ref{Eq:Wienner}
\cite{Book:Haykin02}.

\begin{equation}\label{Eq:Wienner}
    {\bf w}_n^{op}  = {\bf R}^{ - 1} {\bf p}_n
\end{equation}

\noindent onde \textbf{R} \'{e} a matriz de correla\c{c}\~{a}o do vetor de
sinais na entrada do arranjo e \textbf{p} \'{e} o vetor de correla\c{c}\~{a}o
cruzada entre o sinal recebido nas antenas e a sequ\^{e}ncia de
s\'{\i}mbolos desejada do $n$-\'{e}simo usu\'{a}rio (vide equa\c{c}\~{o}es
\ref{Eq:Autocorrelacao} e \ref{Eq:CorelacaoCruzada}).

\begin{equation}\label{Eq:Autocorrelacao}
{\bf R} = E\{{\bf u}(k) {\bf u}(k)^H \},\;\;{\rm onde}\;\;{\bf
u}(k) = {\bf u}(kT)
\end{equation}

\begin{equation}\label{Eq:CorelacaoCruzada}
    {\bf p}_n  = E\{ {\bf u}(k) d_{n}(k)\}
\end{equation}

Finalmente, depois de calcular esses pesos, pode-se usar a equa\c{c}\~{a}o
\ref{Eq:ArrayFactor} para computar o ganho proporcionado pelo
arranjo em uma dada dire\c{c}\~{a}o $\theta_{n}$ (fator do arranjo - vide
equa\c{c}\~{a}o \ref{Eq:Radiation}).

\begin{equation}\label{Eq:ArrayFactor}
    F(\theta _n ) = \left\| {{\bf w}_n^H {\bf f}(\theta _n )} \right\|
\end{equation}

\subsubsection{T\'{e}cnicas de Antenas Inteligentes Investigadas}

As tr\^{e}s arquiteturas de antenas inteligentes consideradas neste
trabalho ser\~{a}o revisadas a seguir. Uma vis\~{a}o geral das antenas
inteligentes e suas aplica\c{c}\~{o}es em sistemas de comunica\c{c}\~{a}o sem fio
pode ser encontrada em \cite{Paper:Godara97, Book:Tsoulos01}.

\begin{description}
    \item[Feixes Fixos Chaveados (FB, do ingl\^{e}s \textit{Fixed Beam})] \'{e} a t\'{e}cnica de antena inteligente
mais simples utilizada nesta disserta\c{c}\~{a}o. Para implementa\c{c}\~{a}o desse
m\'{e}todo \'{e} necess\'{a}rio substituir a antena setorizada de refer\^{e}ncia
por uma matriz de m\'{u}ltiplos feixes mais estreitos em dire\c{c}\~{o}es
preestabelecidas dentro de cada setor da c\'{e}lula. O melhor feixe \'{e}
selecionado de acordo com a posi\c{c}\~{a}o do usu\'{a}rio de forma a
proporcionar maior pot\^{e}ncia ao mesmo. A falha desse m\'{e}todo ocorre
quando um usu\'{a}rio se encontra entre dois l\'{o}bulos adjacentes, onde
o ganho \'{e} baixo.

Na formula\c{c}\~{a}o da t\'{e}cnica de feixes fixos chaveados, simulada nesta
disserta\c{c}\~{a}o, procura-se gerar um n\'{u}mero $L$ de feixes dentro de um
setor de 120 graus em dire\c{c}\~{o}es pr\'{e}-estabelecidas.

Pela equa\c{c}\~{a}o \ref{Eq:Wienner} \'{e} poss\'{\i}vel achar um vetor de pesos,
que aplicado ao arranjo de antenas aponta um feixe na dire\c{c}\~{a}o
$\theta_{n}$. Assim, para direcionar $L$ feixes distribu\'{\i}dos
dentro do setor, pode-se extrapolar a nota\c{c}\~{a}o dada pela equa\c{c}\~{a}o
\ref{Eq:Wienner} e escrever a equa\c{c}\~{a}o \ref{Eq:FixedBeam}.

\begin{equation}\label{Eq:FixedBeam}
    {\bf W} = {\bf R}^{ - 1} {\bf P}
\end{equation}

\noindent onde \textbf{W} \'{e} uma matriz de pesos, onde cada linha \'{e}
composta de um vetor de pesos que direciona um feixe numa dada
dire\c{c}\~{a}o. \textbf{R} e \textbf{P} s\~{a}o dados por \ref{Eq:FBRP}. Note
que \textbf{P} \'{e} constru\'{\i}do de forma a garantir a ortogonalidade
entre os feixes, isto \'{e}, na dire\c{c}\~{a}o do l\'{o}bulo principal de cada
feixe est\~{a}o localizados os nulos dos outros feixes ortogonais
entre si.

\begin{equation}\label{Eq:FBRP}
\begin{array}{l}
 {\bf R} = \left[ {\begin{array}{*{20}c}
   1 & {e^{j\psi_{1,{\theta_1}}}} & {e^{j\psi_{2,{\theta_1}}}} &  \cdots  & {e^{j\psi_{{L-1},{\theta_1}}}}  \\
   1 & {e^{j\psi_{1,{\theta_2}}}} & {e^{j\psi_{2,{\theta_3}}}} &  \cdots  & {e^{j\psi_{{L-1},{\theta_2}}}}  \\
   \vdots  &  \vdots  &  \vdots  &  \ddots  &  \vdots   \\
   1 & {e^{j\psi _{1,{\theta_L}}}} & {e^{j\psi_{2,{\theta_L}}}} &  \cdots  & {e^{j\psi_{{L-1},{\theta_L}}}}  \\
\end{array}} \right]_{LxL}  \\
  \\
 {\bf P} = \left[ {\begin{array}{*{20}c}
   1 & 0 & 0 &  \cdots  & 0  \\
   0 & 1 & 0 &  \cdots  & 0  \\
    \vdots  &  \vdots  &  \vdots  &  \ddots  &  \vdots   \\
   0 & 0 & 0 & 0 & 1  \\
\end{array}} \right]_{LxL}  \\
 \end{array}
\end{equation}

Durante a simula\c{c}\~{a}o \'{e} escolhido o feixe que proporciona maior
ganho para cada usu\'{a}rio, de acordo com a equa\c{c}\~{a}o \ref{Eq:FBGain}.

\begin{equation}\label{Eq:FBGain}
    G_{FB} (\theta _n ) = \mathop {\max }\limits_l [{\bf w}_{l}^H {\bf f}(\theta _n)]
\end{equation}

\noindent onde ${\bf w}_{l}^H$ \'{e} a l-\'{e}sima linha da matriz
\textbf{W} e o operador $\mathop {\max }\limits_l[\cdot]$
representa o valor m\'{a}ximo entre os elementos de um vetor,
considerando toda faixa de valores de $l$.

A figura \ref{Fig:FBPattern4} mostra o padr\~{a}o de radia\c{c}\~{a}o oriundo
da implementa\c{c}\~{a}o da t\'{e}cnica de antena inteligente por feixes fixos
chaveados, usando quatro feixes em um setor de 120 graus. A figura
\ref{Fig:FBPattern8} mostra o mesmo diagrama para oito antenas no
arranjo.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.3]{FBPattern4.eps}
\end{center}
\vspace{-4ex} \caption{\label{Fig:FBPattern4}\hspace{-0.4em}{\bf:}
Antenas inteligentes por feixes fixos chaveados com quatro
elementos no arranjo. Vis\~{a}o polar e cartesiana de feixes fixos em
-45, -15, 15 e 45 graus, dentro de um setor de 120 graus.}
\end{figure}

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.3]{FBPattern8.eps}
\end{center}
\vspace{-4ex} \caption{\label{Fig:FBPattern8}\hspace{-0.4em}{\bf:}
Padr\~{a}o de radia\c{c}\~{a}o das antenas inteligentes por feixes fixos
chaveados com oito elementos no arranjo. Vis\~{a}o polar e cartesiana
de feixes fixos em -52.5, -37.5, -22.5, -7.5, 7.5, 22.5, 37.5 e
52.5 graus, dentro de um setor de 120 graus.}
\end{figure}

\item[Filtro Casado Espacial (MF, do ingl\^{e}s \textit{Matched
Filter})] direciona um feixe para o usu\'{a}rio desejado. O feixe
direcionado \'{e} gerado de acordo com a dire\c{c}\~{a}o do usu\'{a}rio desejado
sem levar em considera\c{c}\~{a}o a posi\c{c}\~{a}o dos interferentes. Dessa
forma, essa t\'{e}cnica almeja maximizar a energia do padr\~{a}o de
radia\c{c}\~{a}o na dire\c{c}\~{a}o do usu\'{a}rio. A figura \ref{Fig:MFConception}
mostra a id\'{e}ia conceitual por tr\'{a}s da t\'{e}cnica de filtro casado
espacial, na qual um feixe direcional acompanha a movimenta\c{c}\~{a}o do
usu\'{a}rio.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.4]{MFConception.eps}
\end{center}
\vspace{-5ex}
\caption{\label{Fig:MFConception}\hspace{-0.4em}{\bf:} Antenas
inteligentes por Filtragem Casada Espacial: um feixe direcional
acompanha a movimenta\c{c}\~{a}o do usu\'{a}rio.}
\end{figure}

Considerando as equa\c{c}\~{o}es \ref{Eq:ArrayFactor} e
\ref{Eq:ArrayReponse}, pode-se escrever o fator do arranjo para
uma estrutura ULA como mostra a equa\c{c}\~{a}o \ref{Eq:ArrayFactor2}.

\begin{equation}\label{Eq:ArrayFactor2}
    F(\theta _n ) = \left\| {{\bf w}^H [\begin{array}{*{20}c}
    1 & {e^{j\psi_{1,\theta _n}}} & {e^{j\psi_{2,\theta_n}}} &  \cdots  & {e^{j\psi_{{M-1},\theta_n}}} \\
\end{array}]^T } \right\|
\end{equation}

\noindent onde $M$ \'{e} o n\'{u}mero de elementos no arranjo.

\'{E} f\'{a}cil notar que para maximizar $F(\theta_n)$ na dire\c{c}\~{a}o de
$\theta_n$, basta considerar o vetor de pesos mostrado na equa\c{c}\~{a}o
\ref{Eq:MFwEquation}.

\begin{equation}\label{Eq:MFwEquation}
    {\bf w_{n}} = [\begin{array}{*{20}c}
    1 & {e^{-j\psi_{1,\theta_n}}} & {e^{-j\psi_{2,\theta_n}}} &  \cdots  & {e^{-j\psi_{{M-1},\theta_n}}} \\
\end{array}]^H  = {\bf f}^*(\theta _n )
\end{equation}

Em \cite{PhD:Rodrigo99} \'{e} mostrado que essa formula\c{c}\~{a}o pode ser
obtida, a menos de um fator de escala, atrav\'{e}s da solu\c{c}\~{a}o MMSE
(\textit{minimum mean square error}) apresentada pela equa\c{c}\~{a}o
\ref{Eq:Wienner}, quando se considera aus\^{e}ncia de desvanecimento.
Para isso, ao desconsiderar o ru\'{\i}do, a matriz de autocorrela\c{c}\~{a}o
\textbf{R} e o vetor de correla\c{c}\~{a}o cruzada \textbf{p}, podem ser
escritos como mostra a equa\c{c}\~{a}o \ref{Eq:MFRP}.

\begin{equation}\label{Eq:MFRP}
\begin{array}{l}
 {\bf R} = {\bf h}{\bf h}^H = \left[ {\begin{array}{*{20}c}
   1 & {e^{-j\psi_{1,{\theta_n}}}} & {e^{-j\psi_{2,{\theta_n}}}} &  \cdots  & {e^{-j\psi_{{M-1},{\theta_n}}}} \\
   {e^{j\psi_{1,{\theta_n}}}} & 1 & {e^{-j\psi_{1,{\theta_n}}}} &  \cdots  & {e^{-j\psi_{{M-2},{\theta_n}}}}  \\
   \vdots  &  \vdots  &  \vdots  &  \ddots  &  \vdots   \\
   {e^{j\psi_{{M-1},{\theta_n}}}} & {e^{j\psi_{{M-2},{\theta_n}}}} & {e^{j\psi_{{M-3},{\theta_n}}}} &  \cdots  & 1  \\
\end{array}} \right]_{MxM}  \\
  \\
 {\bf p_{n}} = {\bf h} = [\begin{array}{*{20}c}
   1 & {e^{j\psi_{1,\theta_n}}} & {e^{j\psi_{2,\theta_n}}} &  \cdots  & {e^{j\psi_{{M-1},\theta_n}}}\\
\end{array}]^T \\
 \end{array}
\end{equation}

O filtro casado espacial (equa\c{c}\~{a}o \ref{Eq:MFwEquation}) \'{e} tamb\'{e}m
obtido atrav\'{e}s do crit\'{e}rio de maximiza\c{c}\~{a}o da rela\c{c}\~{a}o sinal-ru\'{\i}do.
Assim, nota-se que, nesse caso, a maximiza\c{c}\~{a}o da SIR \'{e} equivalente
a minimiza\c{c}\~{a}o do erro quadr\'{a}tico m\'{e}dio.

A figura \ref{Fig:MFPattern4} mostra o padr\~{a}o de radia\c{c}\~{a}o oriundo
da implementa\c{c}\~{a}o da t\'{e}cnica de antenas inteligentes por filtro
casado espacial, usando quatro feixes em um setor de 120 graus. A
figura \ref{Fig:MFPattern8} mostra o mesmo diagrama para oito
antenas no arranjo. Comparando essas duas figuras, \'{e} poss\'{\i}vel
notar que uma maior resolu\c{c}\~{a}o espacial \'{e} conseguida com o aumento
do n\'{u}mero de antenas no arranjo.

Deve-se notar que o l\'{o}bulo sim\'{e}trico inerente \`{a} t\'{e}cnica de
direcionamento de feixe n\~{a}o afetar\'{a} os usu\'{a}rios do sistema. Como
foi mostrado na equa\c{c}\~{a}o \ref{Eq:Radiation}, o padr\~{a}o de radia\c{c}\~{a}o
do conjunto de antenas \'{e} formado pela composi\c{c}\~{a}o do padr\~{a}o de
radia\c{c}\~{a}o do arranjo e do padr\~{a}o de radia\c{c}\~{a}o de um elemento do
arranjo. Dessa forma, o padr\~{a}o de radia\c{c}\~{a}o de um elemento do
arranjo (antena tri-setorizada - vide figura \ref{Fig:Pattern01})
``cancela'' o l\'{o}bulo sim\'{e}trico.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.3]{MFPattern4.eps}
\end{center}
\vspace{-3ex} \caption{\label{Fig:MFPattern4}\hspace{-0.4em}{\bf:}
Padr\~{a}o de radia\c{c}\~{a}o das antenas inteligentes por filtro casado
espacial com quatro elementos no arranjo, dentro de um setor de
120 graus. Vis\~{a}o polar e cartesiana com o usu\'{a}rio em 10 graus.}
\end{figure}

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.3]{MFPattern8.eps}
\end{center}
\vspace{-3ex} \caption{\label{Fig:MFPattern8}\hspace{-0.4em}{\bf:}
Padr\~{a}o de radia\c{c}\~{a}o das antenas inteligentes por filtro casado
espacial com oito elementos no arranjo, dentro de um setor de 120
graus. Vis\~{a}o polar e cartesiana com o usu\'{a}rio em 10 graus.}
\end{figure}

Em \cite{Book:Rappa01} \'{e} mostrado que para um arranjo de antenas
com M elementos sujeito a um canal AWGN (\textit{Additive White
Gaussian Noise}) na aus\^{e}ncia de interfer\^{e}ncia e multi-percursos,
pode-se alcan\c{c}ar um ganho de SNR igual a $10.Log_{10}(M)$.


\item[Antenas Adaptativas (AA, do ingl\^{e}s \textit{Adaptive
Antennas})] \'{e} o esquema mais sofisticado de antenas inteligentes
tratado aqui. As antenas adaptativas geram um feixe direcionado ao
usu\'{a}rio de acordo com seu DOA, enquanto minimiza a energia na
dire\c{c}\~{a}o dos usu\'{a}rios interferentes. Os n\'{\i}veis relativos de
pot\^{e}ncia de cada usu\'{a}rio tamb\'{e}m s\~{a}o levados em considera\c{c}\~{a}o, de
forma que o ganho dado a um usu\'{a}rio \'{e} proporcional ao seu n\'{\i}vel de
pot\^{e}ncia atual.

Adotando a abordagem radar e tendo em mente que essa t\'{e}cnica de
antenas inteligentes leva em considera\c{c}\~{a}o a interfer\^{e}ncia gerada
pelos outros usu\'{a}rios, o vetor que representa o sinal recebido nas
antenas do arranjo pode ser modelado como na equa\c{c}\~{a}o
\ref{Eq:SinalRecExp} (vide equa\c{c}\~{o}es \ref{Eq:CanalST} e
\ref{Eq:SinalRec}).

\begin{equation}\label{Eq:SinalRecExp}
    {\bf u}(t) = \sum\limits_{n = 1}^N {\alpha _n (t){\bf f}(\theta _n
    )} s_n (t)
\end{equation}

\noindent onde N \'{e} o n\'{u}mero de usu\'{a}rios, ${\bf f}(\theta_n)$ \'{e}
dado na equa\c{c}\~{a}o \ref{Eq:ArrayReponse}, $s_n (t)$ dado na equa\c{c}\~{a}o
\ref{Eq:PulsoTrans},   e $\alpha_n(t) = \beta_n \sqrt{P_n}$.
$\beta_n $ \'{e} o coeficiente complexo do canal do n-\'{e}simo usu\'{a}rio
(que inclui tamb\'{e}m um poss\'{\i}vel desvio de fase na demodula\c{c}\~{a}o
coerente do receptor) e $P_{n}$ \'{e} a pot\^{e}ncia de transmiss\~{a}o do
usu\'{a}rio $n$ \cite{PhD:Rodrigo99}. Desse modo, sendo $M$ o n\'{u}mero
de antenas no arranjo, pode-se usar a solu\c{c}\~{a}o \'{o}tima de Wiener
(equa\c{c}\~{a}o \ref{Eq:Wienner}) para calcular os pesos do arranjo.
Nesse caso, desconsiderando o ru\'{\i}do, a matriz de autocorrela\c{c}\~{a}o e
o vetor de correla\c{c}\~{a}o cruzada s\~{a}o dados na equa\c{c}\~{a}o \ref{Eq:AARP}.

\begin{equation}\label{Eq:AARP}
\begin{array}{l}
 {\bf R} = {\bf \Psi \Psi }^H  \\
  \\
 {\bf p}_n  = {\bf \Psi }_{(n)}^*  \\
 \end{array}
\end{equation}

\noindent onde ${\bf \Psi}$ \'{e} a matriz composta de canal e reposta
do arranjo, dada pela equa\c{c}\~{a}o \ref{Eq:MatrizComp}, e ${\bf \Psi
}_{(n)}$ representa a n-\'{e}sima coluna da matriz ${\bf \Psi}$.


\begin{equation}\label{Eq:MatrizComp}
{\bf \Psi } = \left[ {\begin{array}{*{20}c}
   {\alpha_0f_0(\theta_0)} &  \cdots  & {\alpha_{N-1}f_0 (\theta_{N-1})}  \\
   {\alpha_0f_1(\theta_0)} &  \cdots  & {\alpha_{N-1}f_1 (\theta_{N-1})}  \\
    \vdots  &  \ddots & \vdots \\
   {\alpha_0 f_{M-1}(\theta_0)} &  \cdots  & {\alpha_{N-1}f_{M-1}(\theta_{N-1})}  \\
\end{array}} \right]_{MxN}
\end{equation}


\noindent onde $f_j (\theta _n )$ \'{e} a resposta do arranjo na
j-\'{e}sima antena devido \`{a} dire\c{c}\~{a}o de chegada do n-\'{e}simo usu\'{a}rio.

A figura \ref{Fig:AAConception} ilustra a id\'{e}ia conceitual da
t\'{e}cnica de antenas adaptativas. Nessa ilustra\c{c}\~{a}o s\~{a}o simulados
tr\^{e}s usu\'{a}rios simult\^{a}neos localizados em -30, 0 e 45 graus. A
pot\^{e}ncia desses usu\'{a}rios s\~{a}o 1.03, 1.1, 0.8, respectivamente.
Deve-se observar que um feixe \'{e} apontado para o usu\'{a}rio desejado,
enquanto que um nulo lan\c{c}ado ao usu\'{a}rio interferente. Al\'{e}m dessas
duas a\c{c}\~{o}es, a energia do feixe que \'{e} direcionada ao usu\'{a}rio
desejado \'{e} proporcional a sua pot\^{e}ncia de transmiss\~{a}o.

\begin{figure}[!h]
\begin{minipage}[b]{0.30\linewidth}
\includegraphics[width=\linewidth]{AAConception-30.eps}
\centering \textbf{Usu\'{a}rio a -30$^{o}$}
\end{minipage}\hfill
\begin{minipage}[b]{0.30\linewidth}
\includegraphics[width=\linewidth]{AAConception0.eps}
\centering \textbf{Usu\'{a}rio a 0$^{o}$}
\end{minipage}\hfill
\begin{minipage}[b]{0.30\linewidth}
\includegraphics[width=\linewidth]{AAConception45.eps}
\centering \textbf{Usu\'{a}rio a 45$^{o}$}
\end{minipage}
\caption{\label{Fig:AAConception}\hspace{-0.4em}{\bf:} Diagrama de
antenas adaptativas. Tr\^{e}s usu\'{a}rios localizados em -30, 0 e 45
graus, com pot\^{e}ncias 1.03, 1.1, 0.8, respectivamente.}
\end{figure}

\end{description}

Implementa\c{c}\~{o}es pr\'{a}ticas dessas arquiteturas variam das t\'{e}cnicas de
formata\c{c}\~{a}o de feixes fixos at\'{e} sofisticadas t\'{e}cnicas de formata\c{c}\~{a}o
digital de feixes baseadas em processamento adaptativo de sinais
\cite{Book:Litva96}.

\subsection{Microc\'{e}lulas - Arquitetura Celular Hier\'{a}rquica}
\label{Sec:MicroDescription}

A qualidade de servi\c{c}o dos sistemas WCDMA pode ser degradada
devido ao aumento da popula\c{c}\~{a}o de usu\'{a}rios no sistema e a demanda
por servi\c{c}os avan\c{c}ados com altas taxas de transmiss\~{a}o. Para suprir
esse recente e constante crescimento do n\'{u}mero de usu\'{a}rios, as
c\'{e}lulas t\^{e}m diminu\'{\i}do de tamanho. Nesse contexto, surgem as
microc\'{e}lulas. Microc\'{e}lulas s\~{a}o empregadas em conjunto com as
macroc\'{e}lulas j\'{a} existentes, objetivando ajudar na melhoria de
cobertura e aumento de capacidade. Essa estrutura tamb\'{e}m
possibilita aplicar estrat\'{e}gias de balan\c{c}o de carga entre as
c\'{e}lulas e proporcionar servi\c{c}os para usu\'{a}rios com diferentes
perfis de mobilidade.

Microc\'{e}lulas s\~{a}o c\'{e}lulas relativamente pequenas com cobertura
variando de 100 metros a um quilometro. Sua esta\c{c}\~{a}o-base tem
altura comparada com a altura dos pr\'{e}dios (3-6m). Esse tipo de
antena \'{e} geralmente usada em tr\^{e}s \'{a}reas \cite{Book:Pieter}:

\begin{itemize}
    \item \textbf{\textit{Hot Spots}} - \'{a}reas com alta densidade de usu\'{a}rio ou cobertura pobre;
    \item \textbf{Centro Urbano} - \'{a}rea urbana densa com \'{a}rvores e pr\'{e}dios formando
um tipo de ``labirinto" para o sinal celular;
    \item \textbf{Dentro de Pr\'{e}dios} - \'{a}rea interna aos pr\'{e}dios onde os usu\'{a}rios tem mobilidade restrita.
\end{itemize}

Nesta disserta\c{c}\~{a}o, uma estrutura celular hier\'{a}rquica composta de
um camada macrocelular e outra microcelular \'{e} investigada, usando
uma microc\'{e}lula como alternativa para servir uma pequena \'{a}rea com
alta densidade de tr\'{a}fego e eventualmente com cobertura
deficiente. Adotou-se o pleno compartilhamento de freq\"{u}\^{e}ncia entre
as camadas. Esse fato garante o fator de reuso igual a um, por\'{e}m
n\~{a}o garante interfer\^{e}ncia nula entre camadas. Essa interfer\^{e}ncia \'{e}
gerenciada pela separa\c{c}\~{a}o espacial entre micro e macro c\'{e}lulas.

A figura \ref{Fig:HCSGeometria} mostra a geometria relacionada \`{a}
estrutura hier\'{a}rquica modelada nesta disserta\c{c}\~{a}o. Como foi
comentado anteriormente, a macroc\'{e}lula \'{e} assumida como sendo um
conjunto de tr\^{e}s setores hexagonais iguais com raio $R$. O s\'{\i}tio
microcelular \'{e} composto de apenas um setor hexagonal com raio
$R$/10. A microc\'{e}lula \'{e} posicionada dentro do primeiro setor da
esta\c{c}\~{a}o-base central e localizada a uma dist\^{a}ncia \textit{D} da
mesma.

Adotou-se uma altura t\'{\i}pica de antena microcelular igual a 6 m. As
caracter\'{\i}sticas do aparelho m\'{o}vel foram assumidas como sendo
iguais tanto para macroc\'{e}lulas quanto para microc\'{e}lulas.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.4]{HCSGeometria.eps}
\end{center}
\vspace{-5ex}
\caption{\label{Fig:HCSGeometria}\hspace{-0.4em}{\bf:} Geometria
da Estrutura Celular Hier\'{a}rquica (HCS).}
\end{figure}

A figura \ref{Fig:HCSGeometria} mostra que a antena da microc\'{e}lula
n\~{a}o est\'{a} posicionada no centro do setor hexagonal  pequeno. Ao
inv\'{e}s disso, ela \'{e} posicionada na borda da \'{a}rea da microc\'{e}lula, na
dire\c{c}\~{a}o oposta ao padr\~{a}o de radia\c{c}\~{a}o da antena da macroc\'{e}lula.
Essa orienta\c{c}\~{a}o do padr\~{a}o de radia\c{c}\~{a}o da microc\'{e}lula garante um
melhor ajuste da fronteira da microc\'{e}lula. Esse fato pode ser
observado na figura \ref{Fig:HCSProp}, onde a pot\^{e}ncia de
transmiss\~{a}o requerida \'{e} mostrada em fun\c{c}\~{a}o da dist\^{a}ncia.  Devido
ao modelo microcelular de propaga\c{c}\~{a}o utilizado (vide se\c{c}\~{a}o
\ref{Sec:HCSChannel}), o sinal de uma a esta\c{c}\~{a}o microcelular decai
mais r\'{a}pido que o sinal da esta\c{c}\~{a}o macrocelular. Assim, \'{e} f\'{a}cil
notar que o controle de cobertura e o isolamento espacial s\~{a}o
melhor controlados se posicionarmos a microc\'{e}lula como na figura
\ref{Fig:HCSGeometria}, de forma que o seu padr\~{a}o de radia\c{c}\~{a}o
fique frente a frente com o padr\~{a}o de radia\c{c}\~{a}o da antena
macrocelular, ou seja, quando os pontos de energia radiada m\'{a}xima
da microc\'{e}lula e da macroc\'{e}lula est\~{a}o alinhados.

\begin{figure}[!h]
\begin{center}
\includegraphics[scale=0.3]{HCSProp.eps}
\vspace{-3ex} \caption{\label{Fig:HCSProp}\hspace{-0.4em}{\bf:}
Defini\c{c}\~{a}o de fronteiras de c\'{e}lulas: a pot\^{e}ncia oriunda da
microc\'{e}lula decai mais r\'{a}pido que a da macroc\'{e}lula.}
\end{center}
\vspace{-3ex}
\end{figure}

Na HCS, o estabelecimento da fronteira da microc\'{e}lula \'{e} o ponto de
partida para o crit\'{e}rio de sele\c{c}\~{a}o de usu\'{a}rios. A fronteira entre
as \'{a}reas de c\'{e}lulas (microc\'{e}lula/macroc\'{e}lula) \'{e} definida como o
ponto de equil\'{\i}brio, onde a pot\^{e}ncia do sinal de transmiss\~{a}o
requerida do usu\'{a}rio por ambas as c\'{e}lulas \'{e} a mesma
\cite{Paper:Shapira94}. Ent\~{a}o, considerando um usu\'{a}rio localizado
na borda entre microc\'{e}lula e macroc\'{e}lula (ver ponto ``A" nas
figuras \ref{Fig:HCSGeometria} e \ref{Fig:HCSProp}), a sua
pot\^{e}ncia de sinal transmitida para a macroc\'{e}lula pode ser escrita
como na equa\c{c}\~{a}o \ref{Eq:PotTransMA_A}.

\begin{equation}\label{Eq:PotTransMA_A}
    P_{T_M }  = P_{{\rm Req}_M } \left( {D - 2r} \right)^{n_M }
\end{equation}

\noindent onde $n_M$ \'{e} o expoente de perda de percurso da
macroc\'{e}lula, $P_{{\mathop{\rm Re}\nolimits} q_M}$ \'{e} a pot\^{e}ncia do
sinal requerida pela macroc\'{e}lula, $D$ e $r$ podem ser vistos na
figura \ref{Fig:HCSGeometria}. Similarmente, para microc\'{e}lulas, a
equa\c{c}\~{a}o \ref{Eq:PotTransMA_A} \'{e} modificada e torna-se a equa\c{c}\~{a}o
\ref{Eq:PotTransMI_A}.

\begin{equation}\label{Eq:PotTransMI_A}
    P_{T_\mu  }  = P_{{\rm Req}_\mu  } \left( {2r} \right)^{n_\mu  }
\end{equation}

\noindent onde $n_\mu$ \'{e} o expoente de perda de percurso da
microc\'{e}lula, $P_{{\rm Req}_\mu}$ \'{e} a pot\^{e}ncia do sinal requerida
pela microc\'{e}lula. Seguindo a defini\c{c}\~{a}o de fronteira mencionada
anteriormente ($P_{T_\mu}=P_{T_M }$), pode-se chegar na equa\c{c}\~{a}o de
aloca\c{c}\~{a}o de pot\^{e}ncia da microc\'{e}lula (equa\c{c}\~{a}o \ref{Eq:Pot_MA_MI}).

\begin{equation}\label{Eq:Pot_MA_MI}
    P_{{\rm Req}_\mu  }  = P_{{\rm Req}_M } \frac{{\left( {D - 2r}
    \right)^{n_M } }}{{\left( {2r} \right)^{n_\mu  } }}
\end{equation}

A pot\^{e}ncia requerida para cada usu\'{a}rio \'{e} tamb\'{e}m fun\c{c}\~{a}o da classe
de servi\c{c}o do mesmo, obedecendo a modelagem da t\'{e}cnica VSG. No
esquema aqui simulado, a m\'{\i}nima pot\^{e}ncia de transmiss\~{a}o requerida
do m\'{o}vel por qualquer esta\c{c}\~{a}o-base \'{e} imut\'{a}vel em cada classe de
servi\c{c}o de usu\'{a}rio.

A conex\~{a}o \'{e} estabelecida desde que o m\'{o}vel seja capaz de alcan\c{c}ar
a pot\^{e}ncia requerida, enquanto respeita sua limita\c{c}\~{a}o de pot\^{e}ncia
de transmiss\~{a}o m\'{a}xima. Assim, a pot\^{e}ncia de transmiss\~{a}o do m\'{o}vel \'{e}
calculada de acordo com a equa\c{c}\~{a}o \ref{Eq:PC}, tanto para
macroc\'{e}lula como para microc\'{e}lula. No caso da microc\'{e}lula, usa-se
a equa\c{c}\~{a}o \ref{Eq:Pot_MA_MI} para calcular sua pot\^{e}ncia requerida.
O procedimento determin\'{\i}stico de selecionar as bordas das c\'{e}lulas
determina uma fronteira geogr\'{a}fica m\'{e}dia, enquanto que nas
simula\c{c}\~{o}es propriamente ditas, ambas as bordas da macroc\'{e}lula e da
microc\'{e}lula s\~{a}o din\^{a}micas devido \`{a} varia\c{c}\~{a}o do n\'{\i}vel de pot\^{e}ncia
do sinal transmitido, causada pelos efeitos do sombreamento.

%\section{Figuras de M\'{e}rito e Resultados de Desempenho do
%$WISE_{SIM}$} \label{Sec:MeritFigures}
%
%Utilizando o modelo de simula\c{c}\~{a}o exposto nas se\c{c}\~{o}es anteriores,
%pode-se realizar simula\c{c}\~{o}es sist\^{e}micas, relativamente realistas,
%de um sistema WCDMA. Atrav\'{e}s do $WISE_{SIM}$  o desempenho do
%sistema pode ser avaliado atrav\'{e}s do valor da SIR (rela\c{c}\~{a}o
%sinal-interfer\^{e}ncia) por usu\'{a}rio, mensurado ap\'{o}s desespalhamento
%espectral e o processamento do arranjo de antenas no setor da
%esta\c{c}\~{a}o-base de estudo (vide figura \ref{Fig:GridMASHCS}). A
%medida de SIR foi escolhida por figurar uma grandeza
%representativa da taxa de bits errados (BER), medida essencial
%para a garantia de qualidade de servi\c{c}o a n\'{\i}vel de
%inteligibilidade e taxa de transmiss\~{a}o \'{u}til para usu\'{a}rios desse
%sistema.
%
%Os algoritmos aqui propostos ser\~{a}o avaliados atrav\'{e}s de
%compara\c{c}\~{o}es relativas de desempenho, as quais se concentrar\~{a}o
%principalmente na m\'{e}dia e no d\'{e}cimo percentil da SIR. Essas
%medidas d\~{a}o uma id\'{e}ia de capacidade relativa, tendo em vista que
%valores absolutos de capacidade devem ser medidos atrav\'{e}s de
%simuladores que modelam todos os aspectos reais de uma rede de
%comunica\c{c}\~{a}o m\'{o}vel e portanto s\'{o} v\'{a}lidos nessas condi\c{c}\~{o}es. Assim,
%dentro dos objetivos propostos para este trabalho, essa
%apresenta\c{c}\~{a}o relativa de desempenho \'{e} amplamente apropriada para
%garantir seguran\c{c}a e validade nas conclus\~{o}es qualitativas do
%trabalho.





```
